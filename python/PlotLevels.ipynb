```python
import matplotlib.pyplot as plt
%matplotlib notebook
import xarray as xr
```

```python
with xr.open_dataset('../reduceddata/levels.nc') as dss:
    print(dss)
    for rec in range(10000):
        plt.ioff()
        fig,ax = plt.subplots(2, 1, constrained_layout=False)
        ds = dss.isel(k_level=0, record=rec)
        x = ds.XC[0,: ]/1e3
        y = ds.YC[:, 0]/1e3
        pc=ax[0].pcolormesh(x, y, ds.UVEL, cmap='RdBu_r', vmin=-0.4, vmax=0.4)
        fig.colorbar(pc, ax=ax[0], extend='both', shrink=0.6)
        ax[0].set_aspect(1)
        pc = ax[1].pcolormesh(x, y, ds.THETA, vmin=4, vmax=12)
        fig.colorbar(pc, ax=ax[1], extend='both', shrink=0.6)
        ax[1].set_aspect(1)
        fig.savefig('Step%03d.png'%rec, dpi=100)
        plt.close()
        
plt.ion()
```
